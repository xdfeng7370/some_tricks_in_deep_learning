# Spectral bias in deep learning
Consider the following function:
$$f(x)=\sum_{k=1}^5 sin(2kx)/(2k)$$
Suppose we have a sequence points $\{x_m\}_{m=0}^{N-1}$, the frequency index is $j$, then we can obtain 
$$y[j]=\sum_{n=0}^{N-1}x[n]e^{-2\pi i x\cdot j},
$$
where $i$ is the unit imaginary number.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from torch import nn, autograd, optim
from math import pi
import sobol_seq as sobol
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.ticker as ticker

class Unit(nn.Module):

    def __init__(self, in_N, out_N):
        super(Unit, self).__init__()
        self.in_N = in_N
        self.out_N = out_N
        self.L = nn.Linear(in_N, out_N)

    def forward(self, x):
        x1 = self.L(x)
        x2 = torch.tanh(x1)
        return x2


class NN(nn.Module):

    def __init__(self, in_N, width, depth, out_N):
        super(NN, self).__init__()
        self.width = width
        self.in_N = in_N
        self.out_N = out_N
        self.stack = nn.ModuleList()

        self.stack.append(Unit(in_N, width))

        for i in range(depth):
            self.stack.append(Unit(width, width))

        self.stack.append(nn.Linear(width, out_N))

    def forward(self, x):
        # first layer
        for i in range(len(self.stack)):
            x = self.stack[i](x)
        return x


def weights_init(m):
    if isinstance(m, (nn.Conv2d, nn.Linear)):
        nn.init.xavier_normal_(m.weight)
        nn.init.constant_(m.bias, 0.0)


def function_f(x):
    value = torch.zeros_like(x)
    for k in range(1, 6):
        value = value + torch.sin(2 * k * x) / (2 * k)
    return value


def main():
    x = torch.linspace(-pi, pi, 501).reshape(-1, 1)
    f_value = function_f(x)
    # frequency index: 2, 4, 6, 8, 10
    index = torch.tensor([2, 4, 6, 8, 10]) / (2 * pi)
    freq_real = torch.zeros_like(index)
    freq_img = torch.zeros_like(index)
    for i in range(5):
        id = index[i]
        tmp_real = torch.sum(f_value * torch.cos(-2 * pi * x * id))
        tmp_img = torch.sum(f_value * torch.sin(-2 * pi * x * id))
        freq_real[i] = tmp_real
        freq_img[i] = tmp_img
    reference = torch.sqrt((torch.square(freq_real) + torch.square(freq_img)).float())
    print('reference', reference, '\n')

    epochs = 10000
    iter = 50
    in_N = 1
    width = 40
    depth = 2
    out_N = 1

    print(torch.cuda.is_available())
    device = torch.device('cpu' if torch.cuda.is_available() else 'cpu')
    model1 = NN(in_N, width, depth, out_N).to(device)
    model1.apply(weights_init)

    optimizer = optim.Adam(model1.parameters(), lr=1e-2)
    print(model1)
    StepLR = torch.optim.lr_scheduler.StepLR(optimizer, step_size=1000, gamma=0.8)

    Training_loss, Err_y = [], []
    s = 0
    freq_train_real = torch.zeros([int(epochs / iter), 5])
    freq_train_img = torch.zeros([int(epochs / iter), 5])
    for epoch in range(epochs + 1):

        output = model1(x)
        loss = torch.mean(torch.square(output - f_value))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        StepLR.step()
        Training_loss.append(np.log10(np.abs(loss.detach().numpy())))

        if epoch % iter == iter-1:
            print('epoch:', epoch, 'loss:', loss.item())
            for i in range(5):
                id = index[i]
                tmp_real = torch.sum(output * torch.cos(-2 * pi * x * id))
                tmp_img = torch.sum(output * torch.sin(-2 * pi * x * id))
                freq_train_real[s, i] = tmp_real
                freq_train_img[s, i] = tmp_img
            s = s + 1
    result = torch.sqrt((torch.square(freq_train_real) + torch.square(freq_train_img)).float())
    err = torch.abs(result - reference) / reference
    print('err.shape:', err.shape)
    print(result, reference)
    print(1 - err)
    err = err.detach().numpy()
    fig, ax = plt.subplots()
    h = plt.imshow(1 - err, interpolation='nearest', cmap='jet', extent=[0.5, 5.5, 0, 1],
                   origin='lower', aspect='auto')
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    ax.set_xlabel("Frequency")
    ax.set_ylabel("Iteration number")


    plt.colorbar(h, cax=cax)
    plt.show()
    
    
if __name__ == '__main__':
    main()
```
